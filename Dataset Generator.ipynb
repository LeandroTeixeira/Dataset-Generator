### You have (in km) the average distance from each employee to the work, which direction closely describe it, which transportation means they use and if they got covid or not. You wanna see if there's any relation between these datas

### You also have the person's job, salary and time in the company to see if this data helps taking conclusions in any form

#### Setting variables used for generating data

```python
# Variables for the scenario and their respective weights when relevant

# List of the jobs and their respective occurence in the company
job_list = ["Janitor","Junior Web Developer","Senior Web Developer","Intern",
            "Human Resources","Product Marketing","Jurydical Analist",
            "Business Development","Sales Development"]

weights = [0.1,0.17,0.07,0.15,
          0.1,0.13,0.05,
          0.12,0.11]

# How often each job is performed remotely ( = no transportation required)
home_office_p = [ 0, 0.3, 0.35, 0.25,
                0.2, 0.2, 0.25,
                0.25,0.25]

avg_slry = [1300, 2400, 6000, 1000,
            3200, 2800, 9500, 
            4250, 3600]
STD_S = 0.15

# Average time in the company measured in months for realism purposes
avg_time = [14,20, 48, 6,
           18,22,30,
           26,20]

avg_age = [ 40, 28, 35, 21,
    32, 30, 48,
    35, 32]

STD_AGE = [8, 3.5, 5, 3,
    4, 4, 4,
    3, 4]


region_list = ["North", "Northwest","Northeast","Southeast","South","Southwest","East","West"]

# The distance is drawn from a normal distribution with mean and std defined by the user
MEAN_D = 10
STD_D = 6

#If the person lives in a distance closer than this, they live in the center
CENTER_SIZE  = MEAN_D/2

```

#### Generating and setting the data

```python
#functions used to generate the data

def get_avg_time(x):
    seed = np.random.random()
    # Variables used to set value boundaries. Used both to prevent outliers and senseless data (= negative values)
    MIN = (avg - 2*std)
    MAX = (avg + 2*std)
    if seed < MISSING:
        return None
    avg = avg_time[job_list.index(x)]
    std = STD_S*avg_time[job_list.index(x)]
    time = round(np.random.normal(loc = avg, scale = std),0) 
    if(time < MIN):
        return MIN
    if (time > MAX):
        return MAX
    return time

def get_distance(n):
    miss = np.random.random()
    if miss < MISSING:
        return None
    # Negative values are handled by using their absolute values instead
    return np.abs(np.round(np.random.normal(loc = MEAN_D, scale = STD_D),2));

def get_region(n):
    miss = np.random.random()
    if miss < MISSING or n == None:
        return None
    elif n < CENTER_SIZE:
        return "Center"
    else:
        return np.random.choice(region_list)

#Gets the transportation mean of each worker
#The transportation is defined radomly based on salary and profession.

def get_t (n,j):
    a = 2
    miss = np.random.random()
    if miss < MISSING:
        return None
# Extremely poor people will almost never have cars while richer will more often have it. Intermediate values follow a linear function
    if n <= 500:
        car = 0.02
    elif n>= 5000:
        car = 0.98
    else:
        car = n*0.000253 + 0.02
    
    r = np.random.random()
    if r < home_office_p[job_list.index(j)]:
        return "Home Office"
    elif r <= car:
        return "Private Transportation"
    else:
        return "Public Transportation"

MIN_AGE = 17
MAX_AGE = 68
def get_age(x,n):
    miss = np.random.random()
    if miss < MISSING:
        return None
    
    avg = avg_age[job_list.index(x)]
    std = STD_AGE[job_list.index(x)]
    
    age = round(np.random.normal(loc = avg ,scale = std),0) * city_age_multiplier[city_list.index(n)]
    if(age < MIN_AGE):
        age = MIN_AGE * city_age_multiplier[city_list.index(n)]
    if (age > MAX_AGE):
        age = MAX_AGE * city_age_multiplier[city_list.index(n)]
    if(age > 77):
        age = 77
    if (age < 15):
        age = 15
    return int(age)

# Upper and lower bound for salaries
MAX_SAL = 14999 #Can be any value, but should end with 999
MIN_SAL = 800 

def get_salary(x,n):
    salary = np.random.normal(loc = avg_slry[job_list.index(x)], scale = STD_S*avg_slry[job_list.index(x)])
    salary = round (salary * city_salary_multiplier[city_list.index(n)],2)
    if salary < MIN_SAL:
        return np.random.uniform(low = 0.8, high = 1.2) * MIN_SAL
    elif salary > MAX_SAL:
        return np.random.uniform(low = 0.95,high = 1) * MAX_SAL
    else:
        return salary
```

```python
# Data generation

jobs = np.random.choice(job_list,size = SIZE, p = weights)
distances = list(map(get_distance,np.zeros(SIZE)))
avg_ts = list(map(get_avg_time,jobs))
regions = list(map(get_region,distances))
salaries = list(map(get_salary,jobs,cities))
ages = list(map(get_age ,jobs,cities))
transports = list((map(get_t,salaries,jobs)))
```

#### Calculating COVID risks and setting if the person caught it or not

```python
# Age risks will be determined by intervals with 5 ages each
age_risk = []
STD_ar = 0.25

# Salary risk will be determined by $500 intervals
salary_risk = []
STD_sr = 0.25
# Definition of the age and salary risks. L stands for the mean of the distribution. The STD was defined on the last cell

#The values will be necessarily betwen L more or less twice the standard deviation. This is to prevent values equal or less 
# to zero and also to put a ceiling on how big they can become. They also are centered on the data from the group right
# before, to give a sense of continuity

age_risk.append(np.random.normal(loc = 1.1, scale = 2*STD_ar))

for i in range (1,20):
    l = age_risk[int(i-1)]
    aux = np.random.normal(loc = l, scale = STD_ar)
    if (aux < l-(2*STD_ar)):
        aux = l-(2*STD_ar)
    if (aux > l+(2*STD_ar)):
        aux = l+(2*STD_ar)
    age_risk.append(aux)

salary_risk.append(np.random.normal(loc = l, scale = 2*STD_sr))
for i in range(1,int(MAX_SAL/500)):
    l = salary_risk[int(l-1)]
    aux = np.random.normal(loc = l, scale = STD_sr)
    if (aux < l-(2*STD_sr)):
        aux = l-(2*STD_sr)
    if (aux > l+(2*STD_sr)):
        aux = l+(2*STD_sr)
    salary_risk.append(aux)



```

```python
# A matrix that combines the regions data with the cities data to determine how much more 
# likely is someone living there contracting COVID
city_risk = []
AVG_cr = 1.2
STD_cr = 0.3

# Standard deviation between cities natural infection rate
city_infection = []
AVG_ci = 1.1
STD_ci = 0.3

for c in city_list:
    # a variable used to set each city's multiplier for the city_risk matrix
    local_infection = []
    aux_i = np.random.normal (loc = AVG_ci, scale = STD_ci)
    if (aux_i < AVG_ci-(2*STD_ci)):
        aux_i = AVG_ci-(2*STD_ci)
    if (aux_i > AVG_ci+(2*STD_ci)):
        aux_i = AVG_ci+(2*STD_ci)
    
    city_infection.append(aux_i)

    for i in range (0,len(region_list)-1):
        infection = city_infection[city_list.index(c)]
        local_i = np.random.normal(loc = AVG_cr, scale = STD_cr)
        if (local_i < AVG_cr-(2*STD_cr)):
            local_i = AVG_cr-(2*STD_cr)
        if (local_i > AVG_cr+(2*STD_cr)):
            local_i = AVG_cr+(2*STD_cr)
        local_infection.append(local_i*infection)
    
    # The infection of the central region. The formula can be tweaked as needed
    LOW = 1
    HIGH = 1.3
    center = np.mean(local_infection) * np.random.uniform(LOW,HIGH)
    local_infection.append(center)
    #Last, but not least, appending the city name
    local_infection.append(c)
    city_risk.append(local_infection)
```

```python
# Returns the risk multiplier from the specific region. 
def get_region_risk(i):
    region_r = 0
    r_aux = regions[i]
    if (r_aux == None):
        r_aux = get_region(get_distance(0))
    region_r = city_risk[city_list.index(cities[i])][region_list.index(r_aux)]
    return region_r

def get_transport_risk(i,r):
    transport_risk = 1;
    aux = transports[i]
    if (transports[i] == None):
        aux = get_t (salaries[i],jobs[i])
    if aux == "Private Transportation":
        transport_risk = 1.05 * (1 + city_risk[city_list.index(cities[i])][region_list.index("Center")]/10)
    if aux == "Public Transportation":
        transport_risk = 1 + r/2
    return transport_risk
```

```python
# The total chance of getting COVID
covid_risk = []
# We will be generating random data to fill the missing spots.
# This data won't be registered, but will affect the total chance. Thus, the missing chance must be set to 0 before
MISSING = 0

# Now, calculating the total chance per person
for i in range (0, SIZE ):
    covid_chance = city_infection[city_list.index(cities[i])]
    
    region_r = get_region_risk(i) 
    transport_risk = get_transport_risk(i,region_r);
    aux = ages[i]
    if (aux == None):
        aux = get_age(jobs[i],cities[i])
    age_aux = int(aux/ 5)
    print(age_aux)
    age_r = age_risk[age_aux]
    sal_aux = int((salaries[i]/500)-1)
    sal_r = salary_risk[sal_aux]
        
    covid_chance = np.power(covid_chance,region_r)
    covid_chance = np.power(covid_chance,transport_risk)
    covid_chance = np.power(covid_chance,age_risk[age_aux])
    covid_chance = np.power(covid_chance,salary_risk[sal_aux])/5
    
    chance = region_r*age_r*transport_risk*sal_r*covid_chance
    if chance > 0.99:
        chance = 0.99
    if chance < 0.01:
        chance = 0.01
    covid_risk.append(chance)
    

general_stats = [np.mean(covid_risk),np.std(covid_risk),np.median(covid_risk),np.min(covid_risk),np.max(covid_risk)]
```

```python
# The problem will be available in two version, as a classification and regression problem
got_covid = []
for i in range (0,SIZE):
    got_covid.append(np.random.choice(["Yes","No"],p = [covid_risk[i],1-covid_risk[i]]))
p_db["Caught COVID?"] = got_covid
```
